```python
import pandas as pd
reviews = pd.read_csv("morma.csv")
import seaborn as sns
print reviews.shape
```

```python
sns.countplot(reviews.query('Attack < 40').Total)
```

```python
sns.jointplot(x='Attack', y='HP', data=reviews, kind='kde',color='blue')
```

```python
sns.distplot(reviews['Attack'], bins=100, kde=True)
```

```python
sns.kdeplot(reviews.Attack)
```

```python
sns.distplot(reviews.query('Attack < 300').HP, bins=100, kde=True)
```

```python
df = reviews
g = sns.FacetGrid(df, col="Type 1")
g.map(sns.kdeplot, "Attack")
```

```python
sns.pairplot(df[['Attack', 'HP', 'Defense','Speed']],size = 3,diag_kind = 'kde',plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'})
```

```python
sns.pairplot(df, hue = 'Type 1')
```

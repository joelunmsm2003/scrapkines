```python
from skimage import io
```

```python
image=io.imread("fresas.jpg")/255.0 # imread lee las imagenes con los pixeles codificados como enteros 
# en el rango 0-255. Por eso la convertimos a flotante y en el rango 0-1

print("- Dimensiones de la imagen:")
print(image.shape)
plt.imshow(image,vmin=0,vmax=12)
```

```python
lena_rgba=io.imread("fresas.jpg")/255.0 # imread lee las imagenes con los pixeles codificados como enteros 
# en el rango 0-255. Por eso la convertimos a flotante y en el rango 0-1
plt.imshow(lena_rgba) # no es necesario el vmin/vmax para imágenes a color

print("Dimensiones de la imagen:")
print(lena_rgba.shape)
```

```python
plt.imshow(lena_rgba[:,:,0],vmin=0,vmax=1)
plt.title("Canal Rojo")
plt.figure()
plt.imshow(lena_rgba[:,:,1],vmin=0,vmax=1)
plt.title("Canal Verde")
plt.figure()
plt.imshow(lena_rgba[:,:,2],vmin=0,vmax=1)
plt.title("Canal Azul")
```

```python
import pandas as pd
import seaborn as sns
footballers = pd.read_csv("CompleteDataset.csv", index_col=0)
```

```python
print footballers.columns
```

```python
sns.kdeplot(footballers.Age)
```

```python
sns.jointplot(x='Age', y='Potential', data=footballers, kind='kde',color='blue')
```

```python
df = footballers[footballers['Nationality'].isin(['Peru', 'Argentina','Ecuador','England'])]
g = sns.FacetGrid(df, col="Nationality")
g.map(sns.kdeplot, "Age")
```

```python

g = sns.FacetGrid(df, col="Nationality", col_wrap=6)
g.map(sns.kdeplot, "Overall")
```

```python
sns.pairplot(footballers[['Overall', 'Potential', 'Age','Agility']])
```

```python
sns.pairplot(footballers[['Overall', 'Potential', 'Value']])
```
